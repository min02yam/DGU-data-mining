```python
import numpy as np
import pandas as pd

tic2000=pd.read_csv('C:\\Users\\CK\\데이터마이닝\\tic2000.csv')
```

```python
tic2000
```

```python
X=tic2000[tic2000.columns[:85]]
```

```python
X
```

```python
y=tic2000['CARAVAN']
```

```python
y
```

```python
from imblearn.under_sampling import RandomUnderSampler
from sklearn import neighbors, datasets
undersample = RandomUnderSampler(sampling_strategy='majority',random_state=0)  
X_samp,y_samp = undersample.fit_resample(X,y)
df_samp=pd.concat([X_samp,y_samp],axis=1)
```

```python
df_samp
```

```python
df_samp['CARAVAN'].value_counts()
```

```python
x_data=df_samp[df_samp.columns[:85]]
```

```python
y_data=df_samp['CARAVAN']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3,random_state=777) 
```
